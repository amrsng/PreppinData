```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W02/PD2022W02.csv')
```

```python
df.head()
```

#### Remove unwanted fields

```python
df = df.drop(['gender','Parental Contact Name_1','Parental Contact Name_2','Preferred Contact Employer','Parental Contact'], axis=1)
```

```python
df.head()
```

#### Format pupil's name

```python
df['Pupil Name'] = df['pupil first name'] + ' ' + df['pupil last name']
```

```python
df.head()
```

#### Update birthday year to 2022

```python
# convert to datetime
df['dob'] = pd.to_datetime(df['Date of Birth'])
```

```python
# replace year
df["This Year's Birthday"] = df["dob"].apply(lambda x: x.replace(year=2022))
```

```python
df.head()
```

#### Get weekday, make it Friday if it falls on a weekend

```python
df['birthday day'] = df["This Year's Birthday"].dt.day_name()
```

```python
df.head()
```

```python
def convert_weekday(row):
    if row['birthday day'] == 'Saturday':
        return 'Friday'
    elif row['birthday day'] == 'Sunday':
        return 'Friday'
    else:
        return row['birthday day']
```

```python
df['Cake Needed On'] = df.apply (lambda row: convert_weekday(row), axis=1)
```

```python
df.head()
```

#### Get month of birthday

```python
df['Month'] = df["This Year's Birthday"].dt.month_name()
```

```python
df.head()
```

#### Count birthdays in weekdays in month

```python
# df['BDs per Weekday and Month'] 
df_grouped = df.groupby(['Month','Cake Needed On'], as_index = False).size()
```

```python
df_grouped
```

```python
df = df.merge(df_grouped, on = ['Cake Needed On', 'Month'])
```

```python
# test value against Preppin Data sample output
df.loc[df['pupil first name']=='Cal']
df.loc[df['pupil first name']=='Lola']
```

```python
df.head()
```

#### Remove unwanted fields

```python
df = df.drop(['id','pupil first name','pupil last name','dob','birthday day'], axis=1)
```

#### Rename size column

```python
df = df.rename(columns = {'size':'BDs per Weekday and Month'})
```

```python
df.head()
```

#### Export final outputs

```python
df.to_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W02/PD2022W02_final.csv')
```
