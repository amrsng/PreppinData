### Preppin Data 2022 Week 4

https://preppindata.blogspot.com/2022/01/2022-week-4-prep-school-travel-plans.html

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W04/PD2022W04a.csv')
```

```python
df2 = pd.read_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W04/PD2022W04b.csv')
```

#### Join datasets

```python
df3 = pd.merge(df1, df2, how='inner', on=[df1['id'], df2['Student ID']])
```

```python
df3.head()
```

#### Remove unwanted fields

```python
df3.drop(['key_0','key_1','id','pupil first name','pupil last name','gender','Date of Birth','Parental Contact Name_1',\
          'Parental Contact Name_2','Preferred Contact Employer','Parental Contact'], axis=1, inplace=True)
```

```python
df3.head()
```

#### Transform the table so weekdays and travel choices are separate columns

```python
df4 = pd.melt(df3, id_vars='Student ID', value_vars=['M','Tu','W','Th','F'])
```

```python
df4.head()
```

```python
df4.rename(columns=({'variable':'Weekday','value':'Method of Travel'}), inplace=True)
```

```python
df4.head()
```

#### Correct for spelling mistakes

```python
unique_values = df4['Method of Travel'].unique()
```

```python
sorted(unique_values)
```

```python
df4['Method of Travel'] = df4['Method of Travel'].replace('Bycycle','Bicycle')
df4['Method of Travel'] = df4['Method of Travel'].replace('Carr','Car')
df4['Method of Travel'] = df4['Method of Travel'].replace('Helicopeter','Helicopter')
df4['Method of Travel'] = df4['Method of Travel'].replace('Scoter','Scooter')
df4['Method of Travel'] = df4['Method of Travel'].replace('Scootr','Scooter')
df4['Method of Travel'] = df4['Method of Travel'].replace('WAlk','Walk')
df4['Method of Travel'] = df4['Method of Travel'].replace('Waalk','Walk')
df4['Method of Travel'] = df4['Method of Travel'].replace('Walkk','Walk')
df4['Method of Travel'] = df4['Method of Travel'].replace('Wallk','Walk')
```

```python
sorted(df4['Method of Travel'].unique())
```

#### Create sustainable column

```python
df4['Sustainable?'] = 'Sustainable'

df4['Sustainable?'] = np.where(df4['Method of Travel'] == 'Car', 'Non-Sustainable', df4['Sustainable?'])
df4['Sustainable?'] = np.where(df4['Method of Travel'] == 'Aeroplane', 'Non-Sustainable', df4['Sustainable?'])
df4['Sustainable?'] = np.where(df4['Method of Travel'] == 'Helicopter', 'Non-Sustainable', df4['Sustainable?'])
df4['Sustainable?'] = np.where(df4['Method of Travel'] == 'Van', 'Non-Sustainable', df4['Sustainable?'])
```

```python
df4.head()
```

#### Total students by method of travel

```python
df5 = df4.groupby(['Method of Travel','Weekday']).count()
```

```python
df6 = pd.merge(df4, df5, on=['Weekday','Method of Travel'], how='inner')
```

```python
df6.head()
```

```python
df6.drop('Sustainable?_y', axis=1, inplace=True)
```

```python
df6.rename(columns={'Student ID_y':'Number of Trips', 'Student ID_x':'Student ID', 'Sustainable?_x':'Sustainable?'}, inplace=True)
```

```python
df6.head()
```

#### Work out Trips per Day

```python
df6['Trips per Day'] = 1000
```

```python
df6['% of Trips per Day'] = round(df6['Number of Trips']/df6['Trips per Day'], 2)
```

```python
df6.sort_values('Student ID')
```

#### Remove Student ID column and output smaller table

```python
df6.drop('Student ID', axis=1, inplace=True)
```

```python
df7 = df6.drop_duplicates()
```

```python
df7 = df7.reset_index()
```

```python
df7.drop('index', axis=1, inplace=True)
```

```python
df7.head()
```

```python
df7.to_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W04/PD2022W04_final.csv')
```
