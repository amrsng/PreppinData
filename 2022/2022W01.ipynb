#### Load data

```python
import pandas as pd
```

```python
df = pd.read_csv('PD 2022 Wk 1 Input - Input.csv')
```

```python
df.head()
```

#### Pupil's name

```python
df["Pupil's Name"] = df['pupil last name'] + ', ' + df['pupil first name']
```

```python
df.head()
```

#### Parental Contact name

```python
df['Parental Contact Full Name'] = df['pupil last name'] + ', ' + \
df.apply(lambda x: x['Parental Contact Name_1'] if x['Parental Contact'] == 1 else x['Parental Contact Name_2'], axis=1)
```

```python
df['Preferred Parent First Name'] = \
df.apply(lambda x: x['Parental Contact Name_1'] if x['Parental Contact'] == 1 else x['Parental Contact Name_2'], axis=1)
```

```python
df.head()
```

#### Create email

```python
df['Parental Contact Email Address'] = df['Preferred Parent First Name'] + '.' + df['pupil last name'] + \
'@' + df['Preferred Contact Employer'] + '.com'
```

```python
df.head()
```

#### Academic year

```python
def academic_year(x):
    if x['Date of Birth'] >= pd.to_datetime('2014-09-01'):
        return 1
    elif x['Date of Birth'] <= pd.to_datetime('2014-08-31') and x['Date of Birth'] >= pd.to_datetime('2013-09-01'):
        return 2
    elif x['Date of Birth'] <= pd.to_datetime('2013-08-31') and x['Date of Birth'] >= pd.to_datetime('2012-09-01'):
        return 2
    elif x['Date of Birth'] <= pd.to_datetime('2012-08-31') and x['Date of Birth'] >= pd.to_datetime('2011-09-01'):
        return 2
```

```python
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
```

```python
df['Academic Year'] = df.apply(lambda x: academic_year(x), axis=1)
```

```python
df.head()
```

#### Export

```python
df[['Academic Year',"Pupil's Name",'Parental Contact Full Name','Parental Contact Email Address']]\
.to_csv('2022W01_output.csv', index=False)
```
