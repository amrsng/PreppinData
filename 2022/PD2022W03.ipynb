```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W03/PD2022W03a.csv')
df2 = pd.read_csv('/Users/Amar/Desktop/Preppin_Data/2022/2022W03/PD2022W03b.csv')
```

```python
df1.head()
```

```python
df2.head()
```

#### Join tables together

```python
df3 = pd.merge(df1, df2, how='inner', on=[df1['id'],df2['Student ID']])
```

```python
df3.head()
```

#### Remove unwanted fields

```python
df3 = df3.drop(['Parental Contact Name_1','Parental Contact Name_2','Preferred Contact Employer','Parental Contact', \
                'pupil first name','pupil last name','Date of Birth','id', 'key_0','key_1'], axis=1)
```

```python
df3.head()
```

#### Calculate average score per student

```python
df3['Average Score'] = df3[['Maths','English','Spanish','Science','Art','History','Geography']].mean(axis=1)
```

```python
df3.head()
```

#### Pivot table and rename columns

```python
df3_pivot = df3.set_index(["Student ID", "gender", "Average Score"]) \
    .stack() \
    .reset_index(name='Value') \
    .rename(columns={'Maths':'Geography'})
```

```python
df3_pivot.head()
```

```python
df3_pivot = df3_pivot.rename(columns={'level_2': 'Subject', 'Value': 'Score'})
```

```python
df3_pivot.head()
```

#### Round average score

```python
df3_pivot['Average Score'] = round(df3_pivot['Average Score'], 1)
```

```python
df3_pivot.head()
```

#### Check pass rates

```python
def pass_rate(row):
    if row['Score'] >= 75:
        return 1
    else:
        return 0
```

```python
df3_pivot['Pass/Fail'] = df3_pivot.apply (lambda row: pass_rate(row), axis=1)
```

```python
df3_pivot.head()
```

#### Aggregate results and remove unwanted fields

```python
df3_final = df3_pivot.drop(['level_3', 'Score'], axis=1)
```

```python
df3_final.head()
```

```python
df3_final = df3_final.groupby(['Student ID', 'gender','Average Score'])['Pass/Fail'].sum().reset_index()
```

```python
df3_final.head()
```

```python
df3_final.rename(columns = {'Pass/Fail':'Passed Subjects', 'gender':'Gender', 'Average Score': "Student's Avg Score"}, \
                 inplace = True)
```

```python
df3_final.head()
```

#### Output to Excel

```python
df3_final.to_excel('/Users/Amar/Desktop/Preppin_Data/2022/2022W03/PD2022W03_final.xlsx')
```
