### Preppin' Data 2022 W01

https://preppindata.blogspot.com/2022/01/2022-week-1-prep-school-parental.html

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/Amar/Desktop/Preppin_Data/2022W01/PD2022W01.csv')
```

```python
df.head()
```

#### Step 1 - create full name for pupil

```python
df["Pupil's name"] = df['pupil last name'] + ', ' + df['pupil first name']
```

```python
df.head()
```

#### Step 2 - create parent's full name by contact preference

```python
def parental_name(row):
    """ create a function to select the right contact """
    if row['Parental Contact'] == 1:
        return row['Parental Contact Name_1']
    else:
        return row['Parental Contact Name_2']
```

```python
df['Preferred Contact'] = df.apply (lambda row: parental_name(row), axis=1)
```

```python
df['Parental Contact Full Name'] = df['pupil last name'] + ', ' + df['Preferred Contact']
```

```python
df.head()
```

#### Step 3 - create paren't email address

```python
df['Parental Contact Email Address'] = df['Preferred Contact'] + '.' + df['pupil last name'] \
+ '@' + df['Preferred Contact Employer'] + '.com'
```

```python
df.head()
```

#### Step 4 - create academic year for each pupil

```python
#convert string date to datetime
df['dob'] = pd.to_datetime(df['Date of Birth'])
```

```python
def academic_year(row):
    if row['dob'] >= pd.to_datetime('2014-09-01'):
        return 1
    elif row['dob'] >= pd.to_datetime('2013-09-01') and row['dob'] <= pd.to_datetime('2014-08-31'):
        return 2
    elif row['dob'] >= pd.to_datetime('2012-09-01') and row['dob'] <= pd.to_datetime('2013-08-31'):
        return 3
    elif row['dob'] >= pd.to_datetime('2011-09-01') and row['dob'] <= pd.to_datetime('2012-08-31'):
        return 4
    elif row['dob'] >= pd.to_datetime('2010-09-01') and row['dob'] <= pd.to_datetime('2011-08-31'):
        return 5
```

```python
df['Academic Year'] = df.apply (lambda row: academic_year(row), axis=1)
```

```python
df.head()
```

#### Delete unwanted columns

```python
df = df.drop(['id','pupil first name','pupil last name','gender','Date of Birth','Parental Contact Name_1'], axis=1)
df = df.drop(['Parental Contact Name_2', 'Preferred Contact Employer','Parental Contact','Preferred Contact','dob'], axis=1)
```
